```python
import numpy as np
import pandas as pd
```

```python
# 1. Read the data set and replace dashes with 0 to make sure you can perform arithmetic operations on the data.

# I deleted the first epty row in the excel and then imported the file to make sure that first row is coming as header.  

Laliga_df = pd.read_csv("Laliga.csv")

Laliga_df.replace('-',0, inplace= True)
```

```python
# 2. Print all the teams which have started playing between 1930-1980. (5 points)

Laliga_df['Debut'] = Laliga_df['Debut'].astype(str).str[:4].astype(int)

Laliga_df_new = Laliga_df[Laliga_df['Debut'].between(1930, 1980)].replace('-',0)
print(Laliga_df_new)
```

```python
# 3. Print the list of teams which came Top 5 in terms of points 

# First changing the datatype of this column.
Laliga_df.Points = Laliga_df.Points.astype(int)
print(Laliga_df.nlargest(5, ['Points']) )
```

```python
# Write a function with name “Goal_diff_count” which should return all the teams with their Goal Differences. 
# Using the same function, find the team which has maximum and minimum goal difference. (5 points)
# Goal_diff_count = GoalsFor - GoalsAgainst 


Laliga_df.GoalsFor = Laliga_df.GoalsFor.astype(int)
Laliga_df.GoalsAgainst = Laliga_df.GoalsAgainst.astype(int)

def Goal_diff_count(Laliga_df):
    Laliga_df["Goal_diff_count"] = Laliga_df["GoalsFor"] - Laliga_df["GoalsAgainst"]
    print(Laliga_df)
    print(Laliga_df.nlargest(1, ['Goal_diff_count']))
    print(Laliga_df.nsmallest(1, ['Goal_diff_count']))
    
Goal_diff_count(Laliga_df)
```

```python
# 5. Create a new column with name “Winning Percent” and append it to the data set (5 points)
# Percentage of Winning = (GamesWon / GamesPlayed)*100
# If there are any numerical error, replace it with 0%
# Print the top 5 teams which has the highest Winning percentage

Laliga_df.GamesWon = Laliga_df.GamesWon.astype(int)
Laliga_df.GamesPlayed = Laliga_df.GamesPlayed.astype(int)

Laliga_df["Winning Percent"] = (Laliga_df["GamesWon"] / Laliga_df["GamesPlayed"]) * 100
```

```python
print(Laliga_df.nlargest(5, ['Winning Percent']))
```
